## Analyzing and Visualization

The purpose was to gather insights and create visualizations from our dataset (the resulting dataset after the wrangling phase).

Analyzing the dataset gave some insights based on the relationship between entities (columns) in the dataset. Exploring the relationship between different properties in the dataset showed some insights and patterns such as what guarantees a dog would be rated high, what determines if a post would get a lot of likes (favorite count) and so on.

Note - In other to display visualizations, we need to import some libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme()
```

```python
# Set our dataset
df_master = pd.read_csv("twitter_archive_master.csv")
```

### Insight One

From a statistical overview gotten from the dataset, I saw that there was a direct porportion with how high the number of favorite count can get with an increase in retweet count. This makes a lot of sense between when we retweet, we are ideally performing an action that would ensure the tweet gets seen by different users, some of which are not necessarily followers of the original poster.

```python
df_master.sort_values(by=["retweet_count", 'favorite_count']).plot.line("favorite_count", "retweet_count")
plt.title("Retweet Count against Favorite Count")
plt.xlabel("Favorite Count")
plt.ylabel("Retweet Count")
```

### Insight Two

From the analysis above, 94% of those who added more than one picture of thier dog to the tweet got a rating of 10 and above. Seeing the dog in two or more different view can influence how high they get rated.

```python
df_master[df_master['img_num'] > 1]['rating_numerator'].value_counts().plot(kind="pie")
plt.title("Rating Numerator against number of images ")
```

### Insight Three

Across the dataset, about 25% of the dataset does not contain the name of the dog, which is extracted from the Tweet.
This is because reading the Tweet programmatically, you cannot be 100% sure of what the name of the dog is. Introducing a standard introductory text structure to every tweet would help with this.
